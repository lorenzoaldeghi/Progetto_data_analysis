**FUNZIONE PER MODELLI PREDITTIVI**

```python
def modello(X_train,X_test,Y_train,Y_test, model) :
    model_fit = model.fit(X_train,Y_train)
    Y_pred_test=model_fit.predict(X_test)
    Y_pred_train=model_fit.predict(X_train)
    return Y_pred_test, Y_pred_train
    
 
    
    
    
    rmse_test=sqrt(mean_squared_error(Y_test, Y_pred_test))
    print("Root mean squared error test: %.2f" % rmse_test)
    rmse_train=sqrt(mean_squared_error(Y_train, Y_pred_train))
    print("Root mean squared error train: %.2f" % rmse_train)
 
```

```python
model = LinearRegression()
modello(X_train,X_test,Y_train,Y_test, model)
```
